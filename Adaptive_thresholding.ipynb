```
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
import scipy.ndimage
import skimage.filters
from PIL import Image
from skimage.io import imread
from skimage import util
from skimage.color import rgb2gray, gray2rgb
plt.ion()
```

```
#Input image taken and converted into grayscale with resized into 1024*1024
img = cv2.imread('10.png')
grayscale= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grayscale = cv2.resize(grayscale, (1024,1024)) 
n=21
#Adaptive thresholding with region of size 21 with both mean and gaussian weighted average
th2 = cv2.adaptiveThreshold(grayscale,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,n,2)
th3 = cv2.adaptiveThreshold(grayscale, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,n, 2)
```

```
#Original Image
im = Image.fromarray(grayscale)
im.save("10.png")
```

```
# median filtered Image
median_filtered = scipy.ndimage.median_filter(grayscale, size=3)
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')
plt.title('median filtered image')
```

```
print("Adaptive thresholding by mean: ")
plt.imshow(th2, cmap='gray')
plt.title('grayscale')
plt.axis('off')
```

```
print("Adaptive thresholding by gaussian average: ")
plt.imshow(th3, cmap='gray')
plt.title('grayscale')
plt.axis('off')
```

```
im = Image.fromarray(th2)
im.save("4_1.png")
im = Image.fromarray(th3)
im.save("4_2.png")
```
